## Лексические признаки <a class="anchor" id="g1"></a>

* средняя длина слова;
* длина предложения (максимальная, медианная, средняя, дисперсия);
* наличие и нормированное количество «двоеточий»;
* наличие и нормированное количеств «точек с запятой»;
* наличие и нормированное количество «тире»;
* наличие и нормированное количество дефисов;
* наличие и нормированное количество отличных от «точки» пунктуационных знаков конца предложения («! ? …»);
* нормированное количество слов, не имён собственных и не аббревиатур, с большой буквы не в начале предложения;
* наличие и нормированное количество слов с маленькой буквы в кавычках;
* наличие и нормированное количество коротких слов;
* наличие и нормированное количество скобок.

## Морфологические признаки <a class="anchor" id="g2"></a>

* наличие и нормированное количество прилагательных в краткой форме (морфокод содержит прилагательное + категорию «краткая/полная форма»);
* наличие и нормированное количество прилагательных в сравнительных и превосходных степенях сравнения (морфокод содержит прилагательное + категорию «степень сравнения»; иначе прилагательное + маркирующие степень сравнения признаки: окончания и слова «более-менее»);
* нормированное количество личных местоимённых существительных (морфокод содержит указание на личное местоимение; иначе словарь личных местоимений со склонением по падежам);
* наличие и нормированное количество относительных, вопросительных, неопределённых и отрицательных местоимённых существительных (морфокод содержит указание на соответствующее местоимение);
* нормированное количество частиц (морфокод содержит указание на частицу; иначе словарь частиц);
* нормированное количество междометий (морфокод содержит указание на междометие; иначе словарь междометий);
* нормированное количество союзов (морфокод содержит указание на союз);
* нормирование количество союзов и союзных слов для предложений, длина которых больше десяти;
* нормированное количество появлений частицы «не» (ищутся частицы «не» в тексте статистическим алгоритмом);
* нормированное количество притяжательных местоимённых прилагательных (морфокод содержит указание на местоимённое прилагательной);
* нормированное количество местоимённых прилагательных, а также местоимённых наречий (морфокод содержит указание на местоимённое прилагательное или наречие);

* нормированное количество глаголов (морфокод содержит указание на глагол);
* нормированное количество полузнаменательных глаголов (в тексте содержатся глаголы «быть», «стать» и т. д.);
* нормированное количество числительных (морфокод содержит указание на числительное);
* нормированное количество предлогов (морфокод содержит указание на предлог).

## Синтаксические признаки <a class="anchor" id="g3"></a>

* наличие и нормированное количество причастных оборотов (используются методы разбиения предложения на сегменты и ищутся причастия или прилагательные, являющиеся корнем фрагмента текста, затем проверяется, не содержит ли фрагмент глагола или существительного в Им. П.);
* наличие и нормированное количество прилагательных с зависимыми словами (признак определяется на основе морфосинтаксического анализа и последующего анализа древовидной структуры предложения);
* наличие и нормированное количество деепричастных оборотов (используются методы сегментации текста или ищутся деепричастия, являющиеся корнем фрагмента текста, и смотрят, не содержит ли фрагмент глагола или существительного в Им. П);
* наличие и нормированное вводных слов (используются методы сегментации или словарь вводных);
* наличие и нормированное количество нераспространённых предложений в тексте (выявляются предложения, которые содержат только существительные в именительном падеже, инфинитив, глагол, то есть не содержат второстепенных членов);
* наличие и нормированное количество ошибок подчинительной связи при согласовании (после этапа морфологического анализа выделяются биграммы согласований имён существительных и имён прилагательных или причастий, и выявляются ошибки в согласовании грамматических категорий рода, числа, падежа);
* наличие и нормированное количество «фантомных» вершин в синтаксическом дереве предложения, а также наличие и нормированное количество «фантомных вершин» с синтаксической функцией глагола «быть» (после этапа морфосинтаксического анализа определяются вершины, программно добавленные в дерево для установления синтаксических связей);
* наличие и нормированное количество случаев субстантивации прилагательного (после этапа морфосинтаксического анализа образуются вершины, выполняющие функцию слова-субстантиватора, т.е. того существительного, роль которого играет прилагательное, образовавшее субстантивацию);
* наличие и нормированное количество случаев эллипсиса (после этапа морфосинтаксического анализа образуются вершины, выполняющие функцию слов, которые в результате эллипсиса намеренно опускаются из-за своей незначимости для смысла текста);
* наличие и нормированное количество следующих синтаксических членов предложения: сказуемых, подлежащих, обстоятельств, прямых и косвенных дополнений, согласованных и несогласованных определений, приложений, сочетаний с синтаксической функцией отрицания, сочетаний с синтаксической функцией частицы, сочетаний с синтаксической функцией предлога, комплементов, а также подчинительных и сочинительных союзов (после проведения синтаксического подсчитывается нормированное количество соответствующих членов синтаксической структуры предложения);
* наличие и нормированное количество следующих специфических словосочетаний: модальный и полузнаменательный глаголы, модальный глагол и инфинитив, инфинитив и именная часть речи (признак определяется на основе морфосинтаксического анализа и последующего анализа древовидной структуры предложения);
* наличие и нормированное количество следующих дефектностей синтаксической структуры предложения: два и больше корня древовидной структуры и корень синтаксического дерева, не несущий функцию сказуемого, изолированные вершины синтаксического дерева, а также изолированные вершины синтаксического дерева, выраженные предлогами (признаки определяется на основе морфосинтаксического анализа и последующего анализа древовидной структуры предложения);
* наличие и нормированное количество следующих кореферентных структур: синтаксические кореферентные связи, семантические кореферентные связи, местоимённые существительные с начальной формой «я» в косвенных падежах с не определённой кореферентной связью, местоимённые существительные с начальной формой «ты» или «он» с не определённой кореферентной связью (кореферентные связи выявляются алгоритмом референциального анализа, использующим выходные данные морфосинтаксического анализа);

## Идиосинкразические признаки <a class="anchor" id="g4"></a>

* наличие и нормированное количество орфографических ошибок;
* наличие и нормированное количество грамматических ошибок при словообразовании или несуществующие слова (вроде «энтот»);
* наличие и нормированное количество синтаксических ошибок;
* наличие и нормированное количество предложений, не имеющих синтаксическую структуру;
* наличие и нормированное количество сочетаний «пробел-дефис-пробел»;
* наличие и нормированное количество словоформ, содержащих наиболее употребляемые лексические обороты, выполняющие функцию предлога, наречия или предикатива, союза или союзного слова, частицы (списки лексических оборотов составлены на основе открытых данных, предоставляемых Национальным корпусом русского языка);
* наличие и нормированное количество фразеологизмов в тексте (используется словарь фразеологизмов, составленный на основе открытых данных, предоставляемых Национальным корпусом русского языка);
* наличие и нормированное количество «сдвоенных» пробелов.

## Контентно-специфические признаки <a class="anchor" id="g5"></a>

* наличие и нормированное количество иностранных слов (подсчитывается с помощью статистических алгоритмов иностранные слова);
* наличие и нормированное количество сокращений в тексте;
* наличие и нормированное количество акронимов в тексте.

## Структурные признаки <a class="anchor" id="g6"></a>

* наличие и нормированное количество словосочетаний различного вида подчинительной связи (опуская служебные части речи, ищем биграммы «имя существительное + имя», «глагол или имя + имя», «имя или глагол + инфинитив или сравнительная степень прилагательного, наречение, притяжательное местоимение (его, её, их)»)
* наличие и нормированное количество предложений с прямой речью;
* нормированное количество абзацев;
* отношение объёмов zip-архива к объёму исходного текста;
* отношение объёма tar-архива к объёму исходного текста
